<h3>Matriks Asosiasi WAKTU PENGERJAAN</h3>

```python
# Definisikan waktu pengerjaan masing-masing menu
waktu_pengerjaan = {
    'Nasi Liwet': 30,
    'Ayam Goreng': 20,
    'Es Timun Serut': 10,
    'Ayam Betutu': 40,
    'Es Teh': 5,
    'Es Krim': 15,
    'Bakso': 10,
    'Mie Ayam': 15,
    'Es Jeruk': 5,
    'Gulai Kambing': 30,
    'Nasi Putih': 5,
    'Jus Mangga': 10,
    'Ikan Bakar': 20,
    'Sayur Asem': 10,
    'Es Campur': 10,
    'Kerang Balado': 15,
    'Telur Dadar': 10,
    'Nasi Goreng': 20,
    'Rendang': 30,
    'Es Teh Manis': 5,
    'Sate Ayam': 15,
    'Nasi Uduk': 25,  # tambahkan waktu pengerjaan untuk Nasi Uduk
    'Es Kelapa Muda': 10,  # tambahkan waktu pengerjaan untuk Es Kelapa Muda
    'Sop Buntut': 20,  # tambahkan waktu pengerjaan untuk Sop Buntut
    'Jus Jambu': 10  # tambahkan waktu pengerjaan untuk Jus Jambu
}

# Definisikan menu yang ingin dibuat
menu = list(waktu_pengerjaan.keys())

# Buat matriks asosiasi waktu pengerjaan
matriks = [[0 for _ in range(len(menu))] for _ in range(len(menu))]

for i in range(len(menu)):
    for j in range(len(menu)):
        if i == j:
            matriks[i][j] = waktu_pengerjaan[menu[i]]
        else:
            matriks[i][j] = waktu_pengerjaan[menu[i]] + waktu_pengerjaan[menu[j]]

# Tampilkan matriks asosiasi waktu pengerjaan
print("Matriks Asosiasi Waktu Pengerjaan (Menit)")
print("-----------------------------------------")
for i in range(len(menu)):
    print(f"{menu[i]} | ", end="")
    for j in range(len(menu)):
        print(f"{matriks[i][j]:2d} | ", end="")
    print()

# Contoh penggunaan matriks asosiasi waktu pengerjaan
print("\nContoh Penggunaan Matriks Asosiasi Waktu Pengerjaan")
print("-----------------------------------------------")
for i in range(len(menu)):
    for j in range(i+1, len(menu)):
        print(f"Waktu pengerjaan total untuk membuat {menu[i]} dan {menu[j]}: {matriks[i][j]} menit")
```

```python
# Hitung rata-rata lama pengerjaan
total_waktu_pengerjaan = sum(waktu_pengerjaan.values())
rata_rata_waktu_pengerjaan = total_waktu_pengerjaan / len(waktu_pengerjaan)

print(f"Rata-rata lama pengerjaan: {rata_rata_waktu_pengerjaan:.2f} menit")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score

def preprocess_text(text):
    tokens = [t for t in text.split()]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return ' '.join(tokens)

data = pd.read_csv('bda_test.csv')
data['Menu yang dikuasai'] = data['Menu yang dikuasai'].apply(preprocess_text)

X_train, X_test, y_train, y_test = train_test_split(data['Menu yang dikuasai'], data['Nama'], test_size=0.2, random_state=42)

# Create a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# Fit the vectorizer to the training data and transform both the training and testing data
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

clf = MultinomialNB()
clf.fit(X_train_tfidf, y_train)

y_pred = clf.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.3f}')
```
