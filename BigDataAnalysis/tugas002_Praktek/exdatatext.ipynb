```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Baca data
data = pd.read_csv('bda_test.csv')

# Visualisasi hubungan antara menu dan waktu pengerjaan
sns.boxplot(x='Menu yang dikuasai', y='Lama Pengerjaan (Menit)', data=data)
plt.title('Waktu Pengerjaan Berdasarkan Menu')
plt.xticks(rotation=45)
plt.show()

# Hitung rata-rata waktu pengerjaan per menu
rata_rata_per_menu = data.groupby('Menu yang dikuasai')['Lama Pengerjaan (Menit)'].mean()
print(rata_rata_per_menu)

# Cari Pekerja tercepat untuk menu ayam betutu
koki_tercepat_ayam_betutu = data[data['Menu yang dikuasai'].str.contains('ayam betutu')].groupby('Nama')['Lama Pengerjaan (Menit)'].mean().idxmin()
print(f"Pekerja tercepat untuk ayam betutu adalah: {koki_tercepat_ayam_betutu}")
```

```python
menu_counts = data['Menu yang dikuasai'].value_counts()

plt.pie(menu_counts, labels=menu_counts.index, autopct='%1.1f%%')
plt.title('Proporsi yang paling dikuasai')
plt.show()
```
